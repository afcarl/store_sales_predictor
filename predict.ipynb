# PREDICT FOR TEST DATA PROVIDED

```python
# import dependencies
import pandas as pd
import numpy as np
from keras.models import model_from_json
```

```python
data_test = pd.read_csv('test.csv')
data_test
```

```python
data_features = pd.read_csv('features.csv')
data_features
```

```python
merged = pd.merge(data_test,data_features,how='inner',on=['Store','Date'])
merged['Week'] = pd.to_datetime(merged['Date']).dt.week
merged
```

```python
del merged['Fuel_Price']
del merged['Unemployment']
del merged['Date']
del merged['IsHoliday_x']
merged
```

```python
merged.rename(columns={'IsHoliday_y':'IsHoliday'},inplace=True)
merged
```

```python
merged.IsHoliday = merged.IsHoliday.astype(np.int32)
merged
```

```python
sales = merged['SalesForecast']
del merged['SalesForecast']
merged
```

```python
test_data = merged.as_matrix()
test_data
```

```python
# Load Model
json_file = open('model.json','r')
loaded_model_json = json_file.read()
json_file.close()
model = model_from_json(loaded_model_json)
# Load weights
model.load_weights('model.h5')
```

```python
model.compile(loss='mean_squared_error',optimizer='adam')
predictions = model.predict(test_data,verbose=1)
```

# PREDICTION RESULTS

```python
final_dataframe = pd.read_csv('test.csv')
final_dataframe['SalesForecast'] = predictions 
final_dataframe.to_csv('ans.csv')
final_dataframe
```
