# PREDICT THE SALES FOR A STORE

```python
# import dependencies
import pandas as pd
import numpy as np
```

```python
data_train = pd.read_csv('train.csv')
data_train
```

```python
data_features = pd.read_csv('features.csv')
data_features
```

```python
merged = pd.merge(data_train,data_features,how='inner',on=['Store','Date'])
merged['Week'] = pd.to_datetime(merged['Date']).dt.week
merged
```

```python
del merged['Fuel_Price']
del merged['Unemployment']
del merged['Date']
del merged['IsHoliday_x']
merged
```

```python
merged.rename(columns={'IsHoliday_y':'IsHoliday'},inplace=True)
merged
```

```python
merged.IsHoliday = merged.IsHoliday.astype(np.int32)
merged
```

```python
sales = merged['Sales']
del merged['Sales']
sales = sales.as_matrix()
sales
```

```python
train_data = merged.as_matrix()
train_data
```

```python
from sklearn.model_selection import TimeSeriesSplit,train_test_split
from sklearn.metrics import accuracy_score
from sklearn.ensemble import RandomForestRegressor as rfr
```

```python
X_train,X_test,y_train,y_test = train_test_split(train_data,sales,test_size=0.2,random_state=42)
model = rfr(verbose=1)
model.fit(X_train,y_train)
score1 = model.score(X_test,y_test)
score2 = model.score(X_train,y_train)
```

```python
score1
score2
print("Testing Score : %0.2f \n Training Score : %0.2f" %(score1,score2))
```
